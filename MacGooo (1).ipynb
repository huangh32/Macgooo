```python
world_geo = r'world_countries.json'

threshold_scale = np.linspace(each['total_cases'].min(),
                              each['total_cases'].max(),
                              6, dtype=int)
#threshold_scale = threshold_scale.tolist() 
#threshold_scale[-1] = threshold_scale[-1] + 1 

threshold_scale=[each['total_cases'].min(),100000, 1500000, 3000000,each['total_cases'].max()+1]
world_map = folium.Map(location=[0, 0], zoom_start=2, tiles='Mapbox Bright')
world_map.choropleth(
    geo_data=world_geo,
    data=each,
    columns=['location', 'total_cases'],
    key_on='feature.properties.name',
    threshold_scale=threshold_scale,
    fill_color='YlOrRd', 
    fill_opacity=0.6, 
    line_opacity=0.2,
    legend_name='Total Cases',
    reset=True
)

world_map
```
